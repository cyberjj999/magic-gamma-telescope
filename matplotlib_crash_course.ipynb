## Basic Line Plot
- This is a Line Plot. It displays information as a series of points connected by straight line segments.
- It's ideal for showing trends over a continuous domain or time.

**Example Use Case:**

- Tracking Stock Prices Over Time: Suppose you want to visualize the stock price of a company over a year. The x-axis could represent time (in months) and the y-axis could represent the stock price. The continuous nature of the plot helps show trends, rises, and falls.

```python
# Basic Line Plot - exponential example

# simple linear graph, x,y values like (1, 1), (2, 4), etc.
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y)
plt.show()
```

 ## Title, Labels, and Legends

```python
plt.plot(x, y, label='y = x^2')
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

 ## Different Types of Plots

### Histogram

Description: Shows the distribution of data by forming bins along the range of the data and then drawing bars to show the number of observations in each bin.

**Example Use Case:**

- Age Distribution of Employees in a Company: You could use a histogram to visualize the distribution of ages among employees. The x-axis would represent age ranges (like 20-30, 30-40, etc.), and the y-axis would represent the count of employees in each range. This can help HR to understand the age diversity within the organization.

```python
data = [2,2,2,3,3,4,4,4,4,5]
plt.hist(data, bins=4, color='blue', alpha=0.1, label='test')
plt.show()
```
